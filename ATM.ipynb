```python
database ={
    1231:1804,
    1232:1805,
    1233:1806,
    1234:1807,
    1235:1808,
    1236:1809,
    1237:1810,
}
acnt_number=int(input("Enter your Account number: "))
if acnt_number in database.keys():
    password = int(input("Enter Your Pin:"))
    if password == database[acnt_number]:
        def check_balance(balance):
            print(f"Your current balance is: Rs{balance}")

        def deposit(balance, amount):
            if amount > 0:
                balance += amount
                print(f"Rs{amount} Credited successfully to your account.")
            else:
                print("Enter valid amount to Credit")
            return balance

        def withdraw(balance, amount):
            if amount > balance:
                print("Insufficient funds.")
            elif amount <= 0:
                print("Invalid Ammount!\n Please Enter valid Ammount.")
            else:
                balance -= amount
                print(f"Rs{amount} withdrawn successfully from your account.")
            return balance
    else:
        print("Invalid Password ! Try again Later")
else:
    print("Please Enter valid Account number")



def main():
    balance = 0
    while True:
        print("\nWelcome to the ATM!")
        print("1. Check Balance")
        print("2. Deposit")
        print("3. Withdraw")
        print("4. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            check_balance(balance)
        elif choice == '2':
            amount = float(input("Enter the amount to deposit: "))
            balance = deposit(balance, amount)
        elif choice == '3':
            amount = float(input("Enter the amount to withdraw: "))
            balance = withdraw(balance, amount)
        elif choice == '4':
            print("Keep banking with us! Thank you")
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()


```
